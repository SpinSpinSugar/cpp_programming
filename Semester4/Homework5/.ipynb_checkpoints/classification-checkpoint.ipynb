```python
import matplotlib.pyplot as plt
```

```python
with open('dataset', 'r') as f:
    data = f.read()
data = data.split('\n')
data.pop()
data = list(map(lambda l: l.split(), data))
for point in data:
    point[0] = float(point[0])
    point[1] = float(point[1])
```

```python
xcoords = []
ycoords = []
for point in data:
    xcoords.append(point[0])
    ycoords.append(point[1])
```

```python
plt.scatter(xcoords, ycoords, s = 1)
plt.show()
```
